```python
#Find the no of house owned by Cust ID 1 and 2 other than current living homes.(consider only owned house)Also mapped flag type based on sqft table.
df3=pd.DataFrame({'CustId':[1,1,1,1,1],'home city':['Current','pune','pune','pune','Mumbai'],'home type':['rent','owner','owner','owner','rent'],'sqft':[1000,340,356,1250,17000]})
df3
def type1(sqft):
  if 0 <= sqft <= 400:
    return '1Bhk'
  if 400 <= sqft <= 800:
    return '2Bhk'
  if 800 <= sqft <= 1600:
    return '3Bhk'
  if 1600 <= sqft:
    return 'villa'
type1(1200)
```

```python
df3['type'] = df3['sqft'].apply(type1)
df3
```
